```python
import pandas as pd

# Load your IPL matches dataset (update with your actual filename)
df = pd.read_csv('/content/drive/MyDrive/matches.csv')  # e.g., 'ipl_matches.csv'

# Step 1: Drop rows where toss winner or match winner is missing
df.dropna(subset=['winner', 'toss_decision', 'toss_winner'], inplace=True)

# Step 2: Determine whether team batted or fielded first and won
def get_win_type(row):
    if row['winner'] != row['toss_winner']:
        # If match winner ≠ toss winner → they batted/fielded second
        # We can't infer toss decision → skip
        return None
    return 'bat first wins' if row['toss_decision'] == 'bat' else 'field first wins'

df['win_type'] = df.apply(get_win_type, axis=1)

# Step 3: Filter only rows where win_type is not None
df_filtered = df[df['win_type'].notna()]

# Step 4: Group and count
summary = df_filtered.groupby(['winner', 'win_type']).size().unstack(fill_value=0).reset_index()

# Step 5: Rename columns for clarity
summary.columns.name = None
summary.rename(columns={'winner': 'winner'}, inplace=True)

# Step 6: Display clean table
print(summary)
```

```python
import pandas as pd
from prophet import Prophet
import matplotlib.pyplot as plt
```

```python
# Sample data for Mumbai Indians (replace with actual CSV if needed)
data = {
    "Season": [2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023],
    "Wins": [7, 9, 8, 12, 11, 10, 9, 7, 8, 9, 10, 11, 9, 8, 12, 11]
}

df = pd.DataFrame(data)

# Prophet needs columns named "ds" (date) and "y" (target)
df_prophet = df.rename(columns={"Season": "ds", "Wins": "y"})
df_prophet["ds"] = pd.to_datetime(df_prophet["ds"], format='%Y')  # Convert to datetime
```

```python
# Initialize model
model = Prophet()

# Fit the model
model.fit(df_prophet)

# Future dataframe: next 3 seasons
future = model.make_future_dataframe(periods=3, freq='Y')

# Predict
forecast = model.predict(future)

# Display predictions
forecast[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].tail()
```

```python
# Plot forecast
model.plot(forecast)
plt.title("Mumbai Indians - Forecasted Wins")
plt.xlabel("Season")
plt.ylabel("Wins")
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
```

```python
data = {
    'Team': [
        'Mumbai Indians', 'Chennai Super Kings', 'Kolkata Knight Riders',
        'Royal Challengers Bangalore', 'Rajasthan Royals', 'Punjab Kings',
        'Sunrisers Hyderabad', 'Delhi Capitals', 'Gujarat Titans', 'Lucknow Super Giants'
    ],
    'Bat First Wins': [65, 55, 47, 39, 30, 28, 31, 22, 13, 11],
    'Field First Wins': [79, 85, 83, 73, 31, 60, 56, 28, 17, 15]
}

df = pd.DataFrame(data)

# Create Total Wins column
df["Total Wins"] = df["Bat First Wins"] + df["Field First Wins"]
```

```python
features = df[['Bat First Wins', 'Field First Wins', 'Total Wins']]
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)

# Apply KMeans
kmeans = KMeans(n_clusters=3, random_state=42)
df['Cluster'] = kmeans.fit_predict(scaled_features)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x='Bat First Wins', y='Field First Wins',
    hue='Cluster', data=df, palette='Set2', s=100
)

for i in range(len(df)):
    plt.text(df['Bat First Wins'][i] + 0.5, df['Field First Wins'][i],
             df['Team'][i], fontsize=9)

plt.title("Team Clustering Based on Match Wins")
plt.xlabel("Bat First Wins")
plt.ylabel("Field First Wins")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv("/content/drive/MyDrive/matches (1).csv")  # Replace with your filename

# Show basic info
print("🔍 Initial Data Info:")
print(df.info())

# -----------------------------
# 1. Handle Null Values
# -----------------------------
# Check nulls
print("\n🧼 Null Values Before:")
print(df.isnull().sum())

# Optionally fill or drop
df['umpire1'].fillna('Unknown', inplace=True)
df['umpire2'].fillna('Unknown', inplace=True)
df['method'].fillna('Normal', inplace=True)
df.dropna(subset=['winner'], inplace=True)  # Remove rows with no result

# -----------------------------
# 2. Convert Date Column
# -----------------------------
df['date'] = pd.to_datetime(df['date'])

# -----------------------------
# 3. Feature Engineering
# -----------------------------
# Who batted first
df['first_batting'] = df.apply(lambda row: row['team2'] if row['toss_decision'] == 'field' else row['team1'], axis=1)

# Who won the match? (team1 or team2 or draw)
df['match_result'] = df.apply(lambda row: 'Draw' if row['result'] == 'tie' else (
    'Team1' if row['winner'] == row['team1'] else 'Team2'), axis=1)

# Win margin type
df['win_by'] = df.apply(lambda row: 'Runs' if row['result_margin'] > 0 and row['result'] == 'runs' else (
    'Wickets' if row['result_margin'] > 0 and row['result'] == 'wickets' else 'None'), axis=1)

# Toss impact
df['toss_win_match_win'] = df['toss_winner'] == df['winner']

# -----------------------------
# 4. Final Check
# -----------------------------
print("\n✅ Cleaned Data Info:")
print(df.info())
print("\n🧪 Sample:")
print(df[['date', 'team1', 'team2', 'winner', 'match_result', 'win_by', 'toss_win_match_win']].head())
```
